```python
import numpy as np
from sklearn.linear_model import LinearRegression as lr
import pandas as pd
import matplotlib.pyplot as mat
```

```python
a = pd.read_csv("C:/Users/Rigam/Desktop/Rigam/bodyfat.csv")
print(a.head(12))
m,n= a.shape
a = np.array(a)
print(f"the size of the row is {m} \nthe size of the coloumn is : {n}")
height = a[:,4].reshape(-1,1)
weight = a[:,3].reshape(-1,1)
age = a[:,2].reshape(-1,1)
# for i in range(m):
#     height.append(a[i][4])
#     weight.append(a[i][3])
#     age.append(a[i][2])
# height = np.array(height).reshape(-1,1)
# weight = np.array(weight).reshape(-1,1)
# age = np.array(age)
X = np.hstack((height,weight))
model = lr()
model.fit(X,age)       
```

```python
mat.grid()
mat.scatter(height,age,color = "red")
mat.scatter(weight,age,color = "blue")
slope = model.coef_[0]
intercept = model.coef_
equation = f"y = {slope:.2f}x + {intercept:.2f}"
```

```python
def pred(h,w):
    predicted_age= model.predict(np.array([[h,w]]))[0]
    print(f" the predicted age is as follows : {predicted_age}")
    mat.plot(age, predicted_age, color='red', label="Best Fit Line")
```
